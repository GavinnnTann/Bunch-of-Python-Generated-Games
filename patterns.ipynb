```python
import time
import random

def look_busy(duration=60):
    phrases = [
        "Processing data...",
        "Running analysis...",
        "Compiling report...",
        "Synchronizing files...",
        "Checking system status...",
        "Updating records...",
        "Optimizing performance...",
        "Loading resources...",
        "Connecting to server...",
        "Finalizing tasks..."
    ]
    end_time = time.time() + duration
    while time.time() < end_time:
        print(random.choice(phrases))
        time.sleep(random.uniform(1, 3))

#look_busy(30)  # Looks busy for 30 seconds
```

```python
import tkinter as tk
import threading

def look_busy_tk(duration=3000):
    phrases = [
        "Processing data...",
        "Running analysis...",
        "Compiling report...",
        "Synchronizing files...",
        "Checking system status...",
        "Updating records...",
        "Optimizing performance...",
        "Loading resources...",
        "Connecting to server...",
        "Finalizing tasks..."
    ]
    end_time = time.time() + duration

    def log_process():
        while time.time() < end_time:
            phrase = random.choice(phrases)
            textbox.insert(tk.END, phrase + '\n')
            textbox.see(tk.END)
            time.sleep(random.uniform(1, 3))

    thread = threading.Thread(target=log_process)
    thread.start()

root = tk.Tk()
root.title("Data Optimisation")

textbox = tk.Text(root, width=40, height=15)
textbox.pack(padx=10, pady=10)

start_button = tk.Button(root, text="Start", command=lambda: look_busy_tk(30))
start_button.pack(pady=5)

root.mainloop()
```
